This exercise uses clarifai to query the vector representation of a face and get its dot product. It's believed that two same (or similar) faces have higher dot product (closer to 1), but it also depends on the facial orientation (the angle from which the picture is taken). 

```python
from clarifai.rest import ClarifaiApp
from clarifai.rest import Image as ClImage
```

You'll need to register for a key so I'd just hide mine for the purpose of sharing it. 

```python
app = ClarifaiApp(api_key='')
```

The following is my own face, by the way

```python
#we use face embedding ID
model = app.models.get("d02b4508df58432fbb84e800597b8959")
image = ClImage(file_obj=open('../anzoteh96.github.io/images/IMG_2314.jpg', 'rb'))
```

```python
resp=model.predict([image])
```

```python
myvec = resp['outputs'][0]['data']['regions'][0]['data']['embeddings'][0]['vector']
```

From now on we learn that myvec is taken by (response)['outputs'][0]['data']['regions'][0]['data']['embeddings'][0]['vector']

```python
myvec
```

Now we experiment with the other face given by the current Malaysian Prime Minister, Mahathir Mohammad

```python
image2 = ClImage(url='https://www.worldofbuzz.com/wp-content/uploads/2018/07/tun-mahathir-reveals-how-he-manages-to-stay-sharp-and-healthy-at-93-years-old-world-of-buzz-4.jpg')
```

```python
resp2 = model.predict([image2])
```

```python
hisvec2 = resp2['outputs'][0]['data']['regions'][0]['data']['embeddings'][0]['vector']
hisvec2
```

```python
import numpy as np
my_np = np.array(myvec)
his_np = np.array(hisvec2)
```

Now that we have two faces (mine and Mahathir's) let's see what is the dot product between the two (it's 0.48)

```python
np.dot(my_np, his_np)
```

Now we take another face by Mahathir, but taken somewhat from the side. The two face of his gave a dot product of 0.626.

```python
image3 = ClImage(url='https://www.dreshare.com/wp-content/uploads/2018/05/Mahathir-Bin-Mohamad-Net-worth-How-Rich-is-Malaysias-New-Prime-Minister-Actually.jpg')
resp3 = model.predict([image3])
```

```python
hisvec3 = resp3['outputs'][0]['data']['regions'][0]['data']['embeddings'][0]['vector']
hisvec3
```

```python
his_np3 = np.array(hisvec3)
np.dot(his_np, his_np3)
```

We verify that each vector given by Clarifai is unit length. 

```python
np.dot(his_np, his_np)
```
